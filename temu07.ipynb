```python
# Melakukan Unsupervised Learning
# K-means
#Melakukan clustering k-means pada dataset iris
iris2 <- iris
iris2$Species <- NULL

#kmeans (data fitur, banyak kelas):
(kmeans.result <- kmeans(iris2, 3))

#Menampilkan tabel prediksi k-means
table(aktual=iris$Species, prediksi=kmeans.result$cluster)

#Hasil menunjukkan bahwa cluster kelas setosa dapat dengan mudah
#dipisahkan dari cluster lain, sedangkan cluster versicolor dan
#virginica sedikit banyak tumpang tindih satu sama lain (overlapped)


# Menggunakan 4 dimensi fitur dalam data
# Menggunakan 2 dimensi pertama untuk melihat pengelompokan cluster K-means
# yang didapat
plot(iris2[c("Sepal.Length", "Sepal.Width")], col = kmeans.result$cluster)
points(kmeans.result$centers[, c("Sepal.Length", "Sepal.Width")], col = 1:3, pch = 8, cex=2)

# Membagi setiap kelas/spesies iris menjadi 2
# untuk jadi data fitur dan data uji
fiturSetosa <- iris[1:25,1:4]
testSetosa <- iris[26:50,1:4]
fiturVersicolor <- iris[51:75,1:4]
testVersicolor <- iris[76:100,1:4]
fiturVirginica <- iris[101:125,1:4]
testVirginica <- iris[126:150,1:4]


# Menggabungkan semua baris fitur pada variabel train dan
# gabungan baris uji pada variabel test
train <- rbind(fiturSetosa, fiturVersicolor, fiturVirginica)
test <- rbind(testSetosa, testVersicolor, testVirginica)

# Buat vektor cluster sebagai data kelas cluster
cl <- factor(c(rep("setosa",25), rep("versicolor",25), rep("virginica",25)))



# K-Nearest Neighbor
# Lakukan clusterin k-NN dengan menggunakan package dan fungsi
library(class)

#knn(train, test, cl, k=banyak cluster, prob=TRUE/FALSE)
(knn.result <- knn(train, test, cl, k = 3, prob=TRUE))

# Lihat hasil prediksi
table(knn.result)





#Hierarchical Clustering
# Membuat 40 random data dari iris
datalris <- iris
set.seed(1234)
idx <- sample(1:dim(datalris)[1], 40)
irisSample <- datalris[idx,]


# Menyimpan ke kelasActual untuk nanti dibandingkan dengan hasil clustering
kelasAktual <- irisSample$Species
irisSample$Species <- NULL


#Buat plot hirarki clustering
hc <- hclust(dist(irisSample), method="ave")
plot(hc, hang = -1, labels=iris$Species[idx])

# cut tree into 3 clusters
rect.hclust(hc, k=3)
groups <- cutree(hc, k=3)



# Hitung akurasi klasifikasi klustering hirarki
(tabelPrediksiHirarki <- table(aktual=kelasAktual,
                               prediksi=groups))
akurasiHirarki <-
  sum(diag(tabelPrediksiHirarki)/sum(tabelPrediksiHirarki))
akurasiHirarki


```
