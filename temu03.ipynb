```python
# prompt: Uji coba method string berikut: capitalize(), casefold(), center(), count(), encode(), endswith(), expandtabs(), find(), format(), format_map(), index(), isalnum(), isalpha(), isdecimal(), isdigit(), isidentifier(), is slower(), isnumeric(), isprintable(), isspace(), istitle(), isupper(), join(), ljust(), lower(), lstrip(), makestrans(), partition(), replace(), rfind(), rindex(), rjust()

# capitalize()
text = "hello world"
capitalized_text = text.capitalize()
print(capitalized_text)  # Output: Hello world

# casefold()
text = "Hello World"
casefolded_text = text.casefold()
print(casefolded_text)  # Output: hello world

# center()
text = "Hello"
centered_text = text.center(20)
print(centered_text)  # Output:        Hello

# count()
text = "banana"
count_a = text.count('a')
print(count_a)  # Output: 3

# encode()
text = "Hello"
encoded_text = text.encode()
print(encoded_text)  # Output: b'Hello'

# endswith()
text = "Hello world"
ends_with_world = text.endswith("world")
print(ends_with_world)  # Output: True

# expandtabs()
text = "H\te\tl\tl\to"
expanded_text = text.expandtabs(4)
print(expanded_text)  # Output: H   e   l   l   o

# find()
text = "Hello world"
index_world = text.find("world")
print(index_world)  # Output: 6

# format()
name = "John"
age = 30
formatted_text = "My name is {} and I am {} years old.".format(name, age)
print(formatted_text)  # Output: My name is John and I am 30 years old.

# format_map()
data = {'name': 'John', 'age': 30}
formatted_text = "My name is {name} and I am {age} years old.".format_map(data)
print(formatted_text)  # Output: My name is John and I am 30 years old.

# index()
text = "Hello world"
index_world = text.index("world")
print(index_world)  # Output: 6

# isalnum()
text1 = "Hello123"
text2 = "Hello world"
print(text1.isalnum())  # Output: True
print(text2.isalnum())  # Output: False

# isalpha()
text1 = "Hello"
text2 = "Hello123"
print(text1.isalpha())  # Output: True
print(text2.isalpha())  # Output: False

# isdecimal()
text1 = "123"
text2 = "12.3"
print(text1.isdecimal())  # Output: True
print(text2.isdecimal())  # Output: False

# isdigit()
text1 = "123"
text2 = "12.3"
print(text1.isdigit())  # Output: True
print(text2.isdigit())  # Output: False

# isidentifier()
text1 = "my_variable"
text2 = "123variable"
print(text1.isidentifier())  # Output: True
print(text2.isidentifier())  # Output: False

# islower()
text1 = "hello"
text2 = "Hello"
print(text1.islower())  # Output: True
print(text2.islower())  # Output: False

# isnumeric()
text1 = "123"
text2 = "12.3"
print(text1.isnumeric())  # Output: True
print(text2.isnumeric())  # Output: False

# isprintable()
text1 = "Hello world"
text2 = "Hello\nworld"
print(text1.isprintable())  # Output: True
print(text2.isprintable())  # Output: False

# isspace()
text1 = " "
text2 = "Hello"
print(text1.isspace())  # Output: True
print(text2.isspace())  # Output: False

# istitle()
text1 = "Hello World"
text2 = "hello world"
print(text1.istitle())  # Output: True
print(text2.istitle())  # Output: False



# isupper()
text1 = "HELLO"
text2 = "Hello"
print(text1.isupper())  # Output: True
print(text2.isupper())  # Output: False

# join()
my_list = ["apple", "banana", "orange"]
joined_string = ", ".join(my_list)
print(joined_string)  # Output: apple, banana, orange

# ljust()
text = "Hello"
left_justified_text = text.ljust(20)
print(left_justified_text)  # Output: Hello

# lower()
text = "Hello World"
lowercase_text = text.lower()
print(lowercase_text)  # Output: hello world

# lstrip()
text = "   Hello world   "
stripped_text = text.lstrip()
print(stripped_text)  # Output: Hello world

# maketrans()
table = str.maketrans("aeiou", "12345")
text = "Hello world"
translated_text = text.translate(table)
print(translated_text)  # Output: H2ll4 w4rld

# partition()
text = "Hello world"
partitioned_text = text.partition("world")
print(partitioned_text)  # Output: ('Hello ', 'world', '')

# replace()
text = "Hello world"
replaced_text = text.replace("world", "Python")
print(replaced_text)  # Output: Hello Python

# rfind()
text = "Hello world world"
index_world = text.rfind("world")
print(index_world)  # Output: 12

# rindex()
text = "Hello world world"
index_world = text.rindex("world")
print(index_world)  # Output: 12

# rjust()
text = "Hello"
right_justified_text = text.rjust(20)
print(right_justified_text)  # Output:               Hello
```

```python
# prompt: Uji coba method string berikut: rpartition(), rsplit(), rstrip(), split(), splitlines(), startswith(), strip(), swapcase(), title(), translate(), upper(), zfill()

# rpartition()
text = "Hello world world"
rpartitioned_text = text.rpartition("world")
print(rpartitioned_text)  # Output: ('Hello world ', 'world', '')

# rsplit()
text = "apple, banana, orange"
rsplit_text = text.rsplit(", ")
print(rsplit_text)  # Output: ['apple', 'banana', 'orange']

# rstrip()
text = "   Hello world   "
rstrip_text = text.rstrip()
print(rstrip_text)  # Output:    Hello world

# split()
text = "apple, banana, orange"
split_text = text.split(", ")
print(split_text)  # Output: ['apple', 'banana', 'orange']

# splitlines()
text = "Hello\nworld\nPython"
split_lines = text.splitlines()
print(split_lines)  # Output: ['Hello', 'world', 'Python']

# startswith()
text = "Hello world"
starts_with_hello = text.startswith("Hello")
print(starts_with_hello)  # Output: True

# strip()
text = "   Hello world   "
stripped_text = text.strip()
print(stripped_text)  # Output: Hello world

# swapcase()
text = "Hello World"
swapped_text = text.swapcase()
print(swapped_text)  # Output: hELLO wORLD

# title()
text = "hello world"
title_text = text.title()
print(title_text)  # Output: Hello World

# translate()
table = str.maketrans("aeiou", "12345")
text = "Hello world"
translated_text = text.translate(table)
print(translated_text)  # Output: H2ll4 w4rld

# upper()
text = "Hello world"
uppercase_text = text.upper()
print(uppercase_text)  # Output: HELLO WORLD

# zfill()
text = "50"
filled_text = text.zfill(5)
print(filled_text)  # Output: 00050
```
