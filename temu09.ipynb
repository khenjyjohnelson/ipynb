```python
#Deteksi Outlier pada Variabel Tunggal
# Buat satu variabel tunggal yang isinya 100 nilai random
set.seed(3147)
(x <- rnorm(100))
summary(x)


#Deteksi nilai yang outlier di dalam vektor x
boxplot.stats(x)$out

# Visualisasikan sebaran data dan 4 outlier yang telah dideteksi
boxplot(x)


#Deteksi outlier pada Variabel Ganda
# Buat satu variabel lagi yg isinya 100 nilai random
y <- rnorm(100)
df <- data.frame(x, y)
head(df)
summary(df)

# Deteksi nilai yg outlier variabel ganda 'df'
attach(df)
(a <- which(x %in% boxplot.stats(x)$out))

(b <- which(y %in% boxplot.stats(y)$out))
detach(df)


# Deteksi nilai yg outlier di dalam vektor x AND vektor y
(outlier.list1 <- intersect(a,b))
df[outlier.list1,]

plot(df)
points(df[outlier.list1,], col="red", pch="+",cex=2.5)











# Deteksi nilai yg outlier di dalam vektor x OR vektor y
(outlier.list2<-union(a,b))
df[outlier.list2,]


# Plot nilai outlier vektor x OR vektor y
plot(df)
points(df[outlier.list2,], col="blue",pch="x",cex=2)

# Menghilangkan outlier pada data frame vektor x AND vektor y
boxplot(df)
databersih1 <-df[-outlier.list1,]
boxplot(databersih1)



plot(df)
points(df[outlier.list1,], col="red", pch="+", cex=2.5)
plot(databersih1)




#Menghilangkan outlier pada data frame vektor x OR vektor y
boxplot(df)
databersih2 <- df[-outlier.list2, ]
boxplot(databersih2)


plot(df)
points(df[outlier.list2,], col="blue", pch="x", cex=2)
plot(databersih2)



```

```python
#Deteksi Outlier pada Variabel Tunggal
# Buat satu variabel tunggal yang isinya 100 nilai random
set.seed(3147)
(x <- rnorm(100))
summary(x)


#Deteksi nilai yang outlier di dalam vektor x
boxplot.stats(x)$out

# Visualisasikan sebaran data dan 4 outlier yang telah dideteksi
boxplot(x)


#Deteksi outlier pada Variabel Ganda
# Buat satu variabel lagi yg isinya 100 nilai random
y <- rnorm(100)
df <- data.frame(x, y)
head(df)
summary(df)

# Deteksi nilai yg outlier variabel ganda 'df'
attach(df)
(a <- which(x %in% boxplot.stats(x)$out))

(b <- which(y %in% boxplot.stats(y)$out))
detach(df)


# Deteksi nilai yg outlier di dalam vektor x AND vektor y
(outlier.list1 <- intersect(a,b))
df[outlier.list1,]

plot(df)
points(df[outlier.list1,], col="red", pch="+",cex=2.5)


# Step 1: Bangkitkan data
x <- data.frame(matrix(rnorm(60),
                       ncol=3))
x

# Step 2: Variasikan nilai vektor setiap kolom
x[,1] <- x[,1]-3
x[,3] <- x[,3]+3
x


# Step 3: Hitung nilai mean masing-masing fitur
(rerata=apply(x,2,mean))

# Pengurangan untuk setiap nilai elemen thd mean
(matrixrata<-matrix(rep(rerata,20),ncol=3,byrow=T))

# Hitung nilai mean masing2 fitur
(datacov<-(x-matrixrata))

# Step 4: Hitung matriks kovarians (matcov)
matkov <- (t(datacov)%*%as.matrix(datacov))/(nrow(datacov)-1)


# Step 5: Hitung nilai eigen dan vektor eigen dari matriks kovarian
eigennya <- eigen(matkov)
eigennya$vectors
eigennya$values

# Step 6: Pastikan vektor eigen tersortir berdasarkan nilai eigen terbesar sampe terkecil


# Step 7: Konversi nilai eigen ke dalam bentuk persentase komponen utama
(konversi=eigennya$values/sum(eigennya$values) * 100)

(persentase_ciri=cumsum(konversi/sum(konversi) * 100))


# Step 8: Ambil ciri dengan persentasi ciri 80% dari komponen utama
pc <- 80
idx <- as.integer()
for(i in 1:length(persentase_ciri)){
  if(persentase_ciri[i] >= pc){
    idx <- i
    break}

}


# Step 8: Komponen utama dengan nilai persentasi ciri mendekati dari 80%
pc_80=eigennya$vectors[, 1:idx]
(dataPCA=as.matrix(x)%*%pc_80)
x[,1:2]
```
