```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")

text_to_classify = "The stock market is down today"

candidate_labels = ["technology", "workplace", "productivity", "communication"]

results = classifier(text_to_classify, candidate_labels)

print("Text to Classify:")
print(f"\"{text_to_classify}\"")
print("\nClassification Results:")
for label, score in zip(results['labels'], results['scores']):
    print(f"🔹{label}: {score:.4f}")
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")

masked_text = "With advancements in <mask> technology, the possibilities are endless."

predictions = unmasker(masked_text, top_k=3)

print("🔍 Discover the Top Course Topics:\n")
for idx, prediction in enumerate(predictions, 1):
    print(f"🔹 Option {idx}:")
    print(f"Predicted word: {prediction['token_str'].capitalize()}")
    print(f"Confidence: {prediction['score']:.2%}")
    print("\n" + "-" * 40 + "\n")
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)

text = "Hello! I'm Khenjy Johnelson, an undergraduate student in Software Engineering, at Universal University, Batam."

entities = ner(text)

print("🔍 Named Entity Recognition Results:\n")
for idx, entity in enumerate(entities, 1):
    print(f"🔹 Entity {idx}:")
    print(f"• Entity: {entity['word']}")
    print(f"• Type: {entity['entity_group']}")
    print(f"• Confidence: {entity['score']:.2%}")
    print("\n" + "-" * 40 + "\n")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering")

context = (
    "Galuh is a data scientist at OpenAI who enjoys exploring advancements in AI. "
    "He recently moved to Surabaya and has previously worked at Microsoft in Jakarta."
)
question = "Where does Galuh currently live?"

result = qa_pipeline(question=question, context=context)

print("🤖 Question-Answering Result:")
print(f"📌 Context: {context}")
print(f"❓ Question: {question}")
print(f"✅ Answer: {result['answer']}")
```

# **ANALISIS**


### **Berikut adalah analisis hasil dari code di atas:**

1. Zero-shot-classification:
Zero-shot classification untuk mengklasifikasikan teks tanpa pelatihan tambahan, hanya dengan menyebutkan label yang diinginkan. Dalam kode ini, teks "The stock market is down today" diklasifikasikan ke dalam label seperti "technology", "workplace", "productivity", dan "communication." Metode ini menarik karena fleksibel dan cepat untuk aplikasi real-time tanpa perlu melatih model khusus untuk setiap kategori yang ingin diidentifikasi.

2. Text Generation:
Menggunakan pipeline("text-generation") digunakan untuk membuat teks baru berdasarkan prompt. Prompt "Embark on a journey of innovation, where you’ll learn to harness technology to" dijadikan dasar untuk menghasilkan teks yang sesuai. Sangat cocok untuk pembuatan konten kreatif atau rekomendasi teks otomatis.

3. Fill-mask:
Menggunakan pipeline("fill-mask") berfungsi untuk memprediksi kata yang hilang dalam sebuah kalimat. Dalam contoh ini, teks "With advancements in <mask> technology, the possibilities are endless." memberikan prediksi kata yang mungkin cocok di tempat <mask>, seperti "AI" atau "robotics." Metode ini bermanfaat untuk analisis teks kontekstual atau memberikan rekomendasi kata yang sesuai berdasarkan konteks.

4. Named Entity Recognition (NER):
Menggunakan pipeline("ner") untuk mengenali entitas dalam teks, seperti nama, lokasi, organisasi, dan lain-lain. Dalam analisis teks, NER memungkinkan kita mengekstrak informasi penting secara otomatis, memudahkan untuk melakukan analisis data skala besar yang kaya akan detail-detail entitas. Hal ini berguna dalam bidang seperti analisis berita, ekstraksi data otomatis, dan analisis sentimen, karena membantu memahami siapa dan apa saja yang disebutkan dalam teks.

5. Question Answering:
Menggunakan pipeline("question-answering"), model ini dapat menjawab pertanyaan berdasarkan teks atau konteks yang diberikan. Bagian ini biasanya membutuhkan konteks teks yang lebih panjang, dan setelah itu, pengguna dapat memberikan pertanyaan spesifik untuk dijawab oleh model. Fitur ini menarik untuk pembuatan chatbot, layanan pelanggan otomatis, dan aplikasi pencarian informasi karena mampu memahami konteks dengan baik dan memberikan jawaban yang sesuai.

6. Sentiment Analysis:
Menggunakan pipeline("sentiment-analysis") untuk menentukan sentimen (positif, negatif, atau netral) dari sebuah teks. Kode bertujuan untuk mengklasifikasikan emosi dalam teks, yang berguna untuk aplikasi seperti analisis ulasan produk, feedback pelanggan, atau monitor media sosial. Ini membantu perusahaan memahami opini publik secara otomatis dan dalam skala besar.

7. Summarization:
Pada bagian summarization, pipeline("summarization") digunakan untuk meringkas teks yang panjang menjadi lebih singkat tanpa kehilangan inti informasi. Sangat bermanfaat untuk analisis dokumen atau artikel yang panjang, di mana ringkasan yang efisien dan cepat diperlukan. Dalam dunia informasi yang serba cepat, summarization membantu kita menyerap informasi lebih cepat dan fokus pada inti pembahasan.
