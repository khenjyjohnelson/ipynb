```r
w <- c(5,3,4,4,2)
a1 <- c(4,4,5,3,3)
a2 <- c(3,3,4,2,3)
a3 <- c(5,4,2,2,2)
x <- data.frame(a1,a2,a3)
x <- data.frame(t(x))
n_col <- ncol(x)
n_row <- nrow(x)
r <- data.frame(matrix(nrow=n_row,ncol=n_col))
for (j in 1:n_col){
  r[j] <- x[j]/max(x[j])
}
v <- matrix()
for (i in 1:n_row) {
  v[i] <- sum(r[i,] * w)
}
for (i in 1:n_row){
   if (v[i] == max(v)){
     cat("Alternatif yang dipilih adalah Alternatif", i)
     }
}
```

```r
w <- c(5,3,4,4,2)
a1 <- c(4,4,5,3,3)
a2 <- c(3,3,4,2,3)
a3 <- c(5,4,2,2,2)
x <- data.frame(a1,a2,a3)
x <- data.frame(t(x))

c <- c('cost', 'benefit', 'cost', 'cost', 'cost')

n_col <- ncol(x)
n_row <- nrow(x)
r <- data.frame(matrix(nrow=n_row,ncol=n_col))

for (j in 1:n_col){
    if (c[j] == 'benefit'){
        r[j] <- x[j]/max(x[j])
    } else if (c[j] == 'cost') {
        r[j] <- min(x[j])/x[j]
    }
}


v <- matrix()

for (i in 1:n_row) {
  v[i] <- sum(r[i,] * w)
}

for (i in 1:n_row){
   if (v[i] == max(v)){
     cat("Alternatif yang dipilih adalah Alternatif", i)
     }
}
```
