```r
# Contoh 1

w <- c(5,3,4,4,2)
a1 <- c(4,4,5,3,3)
a2 <- c(3,3,4,2,3)
a3 <- c(5,4,2,2,2)
x <- data.frame(a1,a2,a3)
x <- data.frame(t(x))
n_col <- ncol(x)
n_row <- nrow(x)
r <- data.frame(matrix(nrow=n_row,ncol=n_col))
for (j in 1:n_col){
  r[j] <- x[j]/max(x[j])
}
v <- matrix()
for (i in 1:n_row) {
  v[i] <- sum(r[i,] * w)
}
for (i in 1:n_row){
   if (v[i] == max(v)){
     cat("Alternatif yang dipilih adalah Alternatif", i)
     }
}
```

```r
# Versi bisa input

# Allow user input for the number of criteria and alternatives
repeat {
  num_criteria <- suppressWarnings(as.integer(readline("Enter the number of criteria: ")))
  if (!is.na(num_criteria) && num_criteria > 0) break
  cat("Invalid input. Please enter a positive integer.\n")
}

repeat {
  num_alternatives <- suppressWarnings(as.integer(readline("Enter the number of alternatives: ")))
  if (!is.na(num_alternatives) && num_alternatives > 0) break
  cat("Invalid input. Please enter a positive integer.\n")
}

# Initialize empty lists for weights and criteria type
w <- numeric(num_criteria)
c <- character(num_criteria)

# Get weights from user
cat("Enter weights for each criterion: \n")
for (i in 1:num_criteria) {
  repeat {
    w[i] <- suppressWarnings(as.numeric(readline(paste("Weight for criterion", i, ": "))))
    if (!is.na(w[i]) && w[i] > 0) break
    cat("Invalid input. Please enter a positive number.\n")
  }
}

# Get benefit/cost type from user
cat("Enter 'benefit' or 'cost' for each criterion: \n")
for (i in 1:num_criteria) {
  repeat {
    c[i] <- tolower(readline(paste("Type for criterion", i, "(benefit/cost): ")))
    if (c[i] %in% c("benefit", "cost")) break
    cat("Invalid input. Please enter 'benefit' or 'cost'.\n")
  }
}

# Create an empty dataframe to store alternatives
data_list <- list()

# Get alternative values from user
for (i in 1:num_alternatives) {
  alt_values <- numeric(num_criteria)
  for (j in 1:num_criteria) {
    repeat {
      alt_values[j] <- suppressWarnings(as.numeric(readline(paste("Enter value for Alternative", i, "Criterion", j, ": "))))
      if (!is.na(alt_values[j])) break
      cat("Invalid input. Please enter a numeric value.\n")
    }
  }
  data_list[[paste0("a", i)]] <- alt_values
}

# Convert to dataframe
x <- data.frame(data_list)
x <- data.frame(t(x))

# Normalize the decision matrix
n_col <- ncol(x)
n_row <- nrow(x)
r <- data.frame(matrix(nrow=n_row, ncol=n_col))

for (j in 1:n_col) {
  if (c[j] == 'benefit') {
    r[j] <- x[j] / max(x[j])
  } else if (c[j] == 'cost') {
    r[j] <- min(x[j]) / x[j]
  }
}

# Calculate the final score
v <- numeric(n_row)
for (i in 1:n_row) {
  v[i] <- sum(r[i, ] * w)
}

# Determine the best alternative
for (i in 1:n_row) {
  if (v[i] == max(v)) {
    cat("Alternatif yang dipilih adalah Alternatif", i, "\n")
  }
}
```
