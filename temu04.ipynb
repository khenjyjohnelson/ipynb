### Tugas

Buatlah sebuah program yang memanfaatkan setidaknya 5 fungsi/metode pada list

```python
# prompt: Buatlah sebuah program yang memanfaatkan setidaknya 5 fungsi/metode pada list

# 1. append(): Menambahkan elemen ke akhir list
buah = ["apel", "pisang", "mangga"]
buah.append("jeruk")
print(buah)  # Output: ['apel', 'pisang', 'mangga', 'jeruk']

# 2. insert(): Menyisipkan elemen pada indeks tertentu
buah.insert(1, "semangka")
print(buah)  # Output: ['apel', 'semangka', 'pisang', 'mangga', 'jeruk']

# 3. remove(): Menghapus elemen pertama yang cocok dengan nilai yang diberikan
buah.remove("pisang")
print(buah)  # Output: ['apel', 'semangka', 'mangga', 'jeruk']

# 4. pop(): Menghapus dan mengembalikan elemen pada indeks tertentu
buah_terakhir = buah.pop()
print(buah_terakhir)  # Output: jeruk
print(buah)  # Output: ['apel', 'semangka', 'mangga']

# 5. sort(): Mengurutkan elemen dalam list
angka = [3, 1, 4, 2]
angka.sort()
print(angka)  # Output: [1, 2, 3, 4]
```

Buatlah Sebuah program yang memanfaatkan minimal 5 metode tuple

```python
# prompt: Buatlah Sebuah program yang memanfaatkan minimal 5 metode tuple

# 1. count(): Menghitung jumlah kemunculan suatu elemen dalam tuple
my_tuple = (1, 2, 3, 2, 4, 2)
count_2 = my_tuple.count(2)
print(count_2)  # Output: 3

# 2. index(): Mengembalikan indeks kemunculan pertama suatu elemen dalam tuple
index_3 = my_tuple.index(3)
print(index_3)  # Output: 2

# 3. len(): Mengembalikan jumlah elemen dalam tuple
length = len(my_tuple)
print(length)  # Output: 6

# 4. max(): Mengembalikan elemen terbesar dalam tuple
my_tuple2 = (10, 5, 8, 12, 3)
largest = max(my_tuple2)
print(largest)  # Output: 12

# 5. min(): Mengembalikan elemen terkecil dalam tuple
smallest = min(my_tuple2)
print(smallest)  # Output: 3
```
