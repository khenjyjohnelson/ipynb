```python
# Tugas Praktikum

# Nomor 1
#Perbandingan sebaran data Sepal.Length setiap Spesies

# Plotting Sepal Length
boxplot(Sepal.Length ~ Species, data = iris,
        col = c("red", "green", "blue"),
        main = "Perbandingan Sebaran Sepal Length Setiap Spesies")

# Repeat for the other plots
# Sepal Width
boxplot(Sepal.Width ~ Species, data = iris,
        col = c("red", "green", "blue"),
        main = "Perbandingan Sebaran Sepal Width Setiap Spesies")

# Petal Length
boxplot(Petal.Length ~ Species, data = iris,
        col = c("red", "green", "blue"),
        main = "Perbandingan Sebaran Petal Length Setiap Spesies")

# Petal Width
boxplot(Petal.Width ~ Species, data = iris,
        col = c("red", "green", "blue"),
        main = "Perbandingan Sebaran Petal Width Setiap Spesies")




# Nomor 2
# Hasil analisa perbandingan dari persebaran keempat kombinasi fitur
# 1. Sepal Length:
#    Terdapat perbedaan yang jelas antara spesies dalam hal Sepal Length.
#    Umumnya, Setosa memiliki Sepal Length yang lebih kecil dibandingkan
#    dengan Versicolor dan Virginica, yang memiliki sebaran nilai lebih besar.
#    Virginica memiliki nilai Sepal Length yang lebih luas dan lebih besar
#    dibandingkan dengan Versicolor dan Setosa.
#
# 2. Sepal Width:
#    Sepal Width juga menunjukkan perbedaan antara spesies. Setosa cenderung
#    memiliki Sepal Width yang lebih besar dibandingkan dengan Versicolor dan
#    Virginica. Virginica memiliki Sepal Width yang lebih kecil dibandingkan
#    dengan spesies lainnya.
#
# 3. Petal Length:
#    Terdapat perbedaan yang sangat jelas dalam Petal Length di antara spesies.
#    Setosa memiliki Petal Length yang paling kecil, sementara Versicolor dan
#    Virginica memiliki Petal Length yang lebih besar. Virginica memiliki nilai
#    Petal Length yang lebih tinggi dibandingkan Versicolor.
#
# 4. Petal Width:
#    Mirip dengan Petal Length, Petal Width menunjukkan perbedaan yang signifikan.
#    Setosa memiliki Petal Width yang lebih kecil dibandingkan dengan Versicolor
#    dan Virginica, dengan Virginica menunjukkan nilai Petal Width yang paling besar.



# Nomor 3
# Plot 1: Sepal Length vs. Sepal Width
with(iris, plot(Sepal.Length, Sepal.Width, col=c("red", "green", "blue")[Species],
                pch=as.numeric(Species), main="Sepal Length vs Sepal Width"))
mtext("Red: Setosa    Green: Versicolor    Blue: Virginica", side = 4, line = 0, cex = 0.7, col = "black")

# Plot 2: Petal Length vs. Petal Width
with(iris, plot(Petal.Length, Petal.Width, col=c("red", "green", "blue")[Species],
                pch=as.numeric(Species), main="Petal Length vs Petal Width"))
mtext("Red: Setosa    Green: Versicolor    Blue: Virginica", side = 4, line = 0, cex = 0.7, col = "black")

# Plot 3: Sepal Length vs. Petal Length
with(iris, plot(Sepal.Length, Petal.Length, col=c("red", "green", "blue")[Species],
                pch=as.numeric(Species), main="Sepal Length vs Petal Length"))
mtext("Red: Setosa    Green: Versicolor    Blue: Virginica", side = 4, line = 0, cex = 0.7, col = "black")

# Plot 4: Sepal Width vs. Sepal Length
with(iris, plot(Sepal.Width, Sepal.Length, col=c("red", "green", "blue")[Species],
                pch=as.numeric(Species), main="Sepal Width vs Sepal Length"))
mtext("Red: Setosa    Green: Versicolor    Blue: Virginica", side = 4, line = 0, cex = 0.7, col = "black")



# Nomor 4
# Hasil analisa perbandingan dari persebaran keempat kombinasi fitur
# 1. Sepal Length vs. Sepal Width:
#    Plot ini menunjukkan bahwa Setosa cenderung memiliki Sepal Width yang lebih
#    besar dengan Sepal Length yang lebih kecil. Versicolor dan Virginica memiliki
#    rentang yang lebih luas untuk Sepal Length dan Sepal Width. Setosa seringkali
#    dapat dibedakan dengan jelas dari Versicolor dan Virginica.
#
# 2. Petal Length vs. Petal Width:
#    Plot ini menunjukkan bahwa Setosa memiliki Petal Length dan Petal Width yang
#    lebih kecil dibandingkan dengan Versicolor dan Virginica. Virginica memiliki
#    nilai Petal Length dan Petal Width yang paling besar, sedangkan Versicolor
#    berada di tengah.
#
# 3. Sepal Length vs. Petal Length:
#    Plot ini memperlihatkan bahwa Setosa memiliki Petal Length yang lebih kecil
#    dengan Sepal Length yang lebih kecil juga. Versicolor dan Virginica memiliki
#    variasi yang lebih besar dalam Petal Length, dengan Virginica menunjukkan
#    nilai yang paling tinggi.
#
# 4. Sepal Width vs. Sepal Length:
#    Plot ini menunjukkan bahwa Setosa memiliki Sepal Width yang lebih besar pada
#    rentang Sepal Length yang lebih kecil. Versicolor dan Virginica memiliki nilai
#    Sepal Length dan Sepal Width yang lebih bervariasi. Virginica cenderung
#    menunjukkan Sepal Length yang lebih besar dengan Sepal Width yang lebih kecil.
```
