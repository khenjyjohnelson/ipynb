```python
# isupper()
text1 = "HELLO"
text2 = "Hello"
print(text1.isupper())  # Output: True
print(text2.isupper())  # Output: False


teks = "ENAK SEKALI!"
print(teks.isupper());

teks = "EN4K S3K4L1!"
print(teks. isupper());

teks = "ENAK SeKALI!"
print(teks.isupper());
```

```python
# upper()
text = "Hello world"
uppercase_text = text.upper()
print(uppercase_text)  # Output: HELLO WORLD
```

```python
# lower()
text = "Hello World"
lowercase_text = text.lower()
print(lowercase_text)  # Output: hello world
```

```python
# title()
text = "hello world"
title_text = text.title()
print(title_text)  # Output: Hello World
```

```python
# swapcase()
text = "Hello World"
swapped_text = text.swapcase()
print(swapped_text)  # Output: hELLO wORLD
```

```python
# startswith()
text = "Hello world"
starts_with_hello = text.startswith("Hello")
starts_with_world = text.startswith("world")
print(starts_with_hello)  # Output: True



text = "b a a coba"

# dengan menggunakan start dan end
print(text.startswith("b", 0, 9))
print(text.startswith("a", 2, 9))
print(text.startswith("b", 1, 9))
```

```python
# strip()
text = "   Hello world   "
stripped_text = text.strip()
print(stripped_text)  # Output: Hello world

strip_from_l = stripped_text.strip(' l')
print(strip_from_l)



string = ' lololo kofe lololo '

print(string.strip(' loe')) #melihat karakter pertama yang bakal distrip

print(string.strip('loe')) #tidak ada spasi jadi tidak ada

string = 'python keren'
print(string.strip('py'))
```

```python
# rstrip()
text = "   Hello world   "
rstrip_text = text.rstrip()
print(rstrip_text)  # Output:    Hello world
```

```python
# lstrip()
text = "   Hello world   "
stripped_text = text.lstrip()
print(stripped_text)  # Output: Hello world

stripped_text = text.lstrip()
```

```python
# replace()
text = "Hello world"
replaced_text = text.replace("world", "Python")
print(replaced_text)  # Output: Hello Python
```
