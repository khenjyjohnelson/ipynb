```python
username = (input("Masukkan username: "))
peran = (input("Masukkan peran (admin/editor/viewer): "))
status = input("Masukkan status (aktif/nonaktif): ")
group = input("Masukkan group (teknologi/pemasaran/keuangan): ")

#Operasi logika

if (peran == "admin"):
  print("Akses Penuh diberikan", username)
elif (peran == "editor"):
  print("Akses terbatas diberikan", username)
elif (peran == "viewer"):
  print("Akses ditolak, hak tidak mencukupi,", username)
else :
  print("Akses ditolak, peran tidak valid,", username)

#Operasi identitas
if(status == "aktif"):
  print("Akses berhasil, ", username)
elif (status == "nonaktif"):
  print("Akun Anda nonaktif, akses ditolak", username)

#Operasi keanggotaan
if (group in ("teknologi, pemasaran, keuangan")):
  print("Akses ditolak, ", username)
elif (group in ("pemasaran")):
  print("Akses ditolak, ", username)
elif (group in ("keuangan")):
  print("Akses ditolak, ", username)
```

```python
 prompt: Buatlah sebuah program yang memeriksa
# apakah sebuah angka adalah angka genap atau
# ganjil
# Buatlah sebuah program yang menentukan
# grade (A, AB, B, BC, C, D, E) berdasarkan nilai.
# (Sesuaikan dengan penilaian di Uvers)
# Buatlah sebuah program dengan
# menggunakan nested if

# Program untuk memeriksa apakah sebuah angka genap atau ganjil
angka = int(input("Masukkan sebuah angka: "))
if angka % 2 == 0:
  print(angka, "adalah angka genap.")
else:
  print(angka, "adalah angka ganjil.")

# Program untuk menentukan grade berdasarkan nilai
nilai = int(input("Masukkan nilai: "))
if nilai >= 85:
  grade = "A"
elif nilai >= 80:
  grade = "AB"
elif nilai >= 75:
  grade = "B"
elif nilai >= 70:
  grade = "BC"
elif nilai >= 65:
  grade = "C"
elif nilai >= 60:
  grade = "D"
else:
  grade = "E"
print("Grade Anda adalah:", grade)

# Program dengan nested if
umur = int(input("Masukkan umur Anda: "))
jenis_kelamin = input("Masukkan jenis kelamin Anda (L/P): ")
if umur >= 18:
  if jenis_kelamin == "L":
    print("Anda adalah pria dewasa.")
  else:
    print("Anda adalah wanita dewasa.")
else:
  if jenis_kelamin == "L":
    print("Anda adalah anak laki-laki.")
  else:
    print("Anda adalah anak perempuan.")
```

### Tugas Perulangan

looping for atau while
Buatlah sebuah program yang menggunakan
break atau continue

```python
# prompt: Buatlah sebuah program yang menggunakan
# looping for atau while
# Buatlah sebuah program yang menggunakan
# break atau continue

# Program menggunakan looping for
for i in range(1, 11):
  print(i)

# Program menggunakan looping while
count = 1
while count <= 5:
  print(count)
  count += 1

# Program menggunakan break
for i in range(1, 11):
  if i == 6:
    break
  print(i)

# Program menggunakan continue
for i in range(1, 11):
  if i % 2 == 0:
    continue
  print(i)
```
