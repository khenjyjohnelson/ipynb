```python
library(MASS)
library(e1071)

#Inisiasi Dataset Iris
data(iris)

train<-sample(1:150,50)

test<-iris[-train,]

summary(test$Species)






#Buat Model Prediksi LDA
#Menggunakan data training
modelLDA<-lda(Species~.,iris,subset=train)

#Cek model prediksi LDA
modelLDA
modelLDA$counts
modelLDA$means
modelLDA$scaling


#Prediksi Model LDA
predLDA=predict(modelLDA,test)

#Cek hasil output, posterior, dan proyek liner diskriminasinya
summary(predLDA$class)
head(predLDA$posterior,3) # posterior prob
head(predLDA$x,3) # LD projections


#Hasil prediksi LDA
#Confusion Matriks
(LDATable <- table(prediksi=predLDA$class, aktual=test$Species))

#Akurasi LDA
(AkurasiLDA <- sum(diag(LDATable))/sum(LDATable))








#Buat Model Prediksi SVM
#Membuat model training SVM menggunakan data training
modelSVM <- svm(Species ~. , iris, kernel="linear",subset = train)

#Cek support vector dari parameter kernel
modelSVM$index
summary(modelSVM)


#Prediksi Model SVM
SVMpred<-predict(modelSVM, test)

#Hasil Prediksi SVM
#Confusion Matriks
(SVMTable <- table(prediksi=SVMpred, aktual=test$Species))

#Akurasi LDA
(AkurasiSVM <- sum(diag(SVMTable))/sum(SVMTable))






#Membuat Model Prediksi RBF
#Buat Model Training
modelRBF <- svm(Species ~ ., iris, kernel="radial", subset = train)

#Cek model prediksi RBF
modelRBF$index
summary(modelRBF)

#Prediksi model dengan data uji
RBFpred <- predict(modeIRBF, test)

#Confusion matriks
(RBFTable <- table(prediksi=RBFpred, aktual=test$Species))

#Akurasi RBF
(AkurasiRBF <- sum(diag(RBFTable))/sum(RBFTable))


paste("Akurasi LDA:", AkurasiLDA*100, "%,",
      "Akurasi SVM:", AkurasiSVM*100, "%,",
      "Akurasi RBF:", AkurasiRBF*100, "%", sep=" ")

```
